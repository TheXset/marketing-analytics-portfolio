```python
import pandas as pd
launch = pd.read_csv('/content/launch_test_task.csv', sep=';')
launch.head()
```

```python
print(f'Количество строк: {launch.shape[0]}')
print(f'Количество столбцов: {launch.shape[1]}')
```

```python
launch['platform'].value_counts()
```

```python
launch['app_platform'] = launch['platform'].apply(
    lambda x: 'Android' if 'Android' in x else 'iOS'
)

grouped = (
    launch
    .groupby(['source', 'app_platform'])
    .agg(
        unique_users=('user_id', 'nunique'),
        sessions=('session_id', 'count')
    )
    .unstack(fill_value=0)
)

grouped.columns = [f'{metric}_{platform}'
                   for metric, platform in grouped.columns]

task1 = grouped.reset_index()
task1 = (
    task1
    .assign(total=task1['unique_users_Android'] + task1['unique_users_iOS'])
    .sort_values('total', ascending=False)
    .drop(columns='total')
)

task1.to_csv('task1_source_platform.csv', index=False)
task1.head(10)
```

```python
task2 = (
    launch
    .groupby('region')
    .agg(unique_users=('user_id', 'nunique'))
    .reset_index()
    .sort_values('unique_users', ascending=False)
    .rename(columns={'region': 'city'})
)

task2.to_csv('task2_users_by_city.csv', index=False)
task2.head(10)
```
